```python
import random
from sympy import *
from scipy.optimize import fsolve

def generate_combined_expression_question(complexity=0.5, variability=0.5):
    x = symbols('x')
    num_functions = random.randint(2, 4)
    functions = []

    for _ in range(num_functions):
        function_type = random.choices(['polynomial', 'trigonometric', 'exponential', 'logarithmic'], weights=[0.4, 0.3, 0.2, 0.1])[0]
        if function_type == 'polynomial':
            degree = random.randint(2, 3)
            coeffs = [random.randint(1, 5) for _ in range(degree)]
            function = sum(coeff * x**i for i, coeff in enumerate(coeffs))
        elif function_type == 'trigonometric':
            function = random.choice([sin(x), cos(x), tan(x)])
        elif function_type == 'exponential':
            function = random.choice([exp(x), exp(-x)])
        elif function_type == 'logarithmic':
            function = random.choice([log(x), log(x + 1)])
        
        functions.append(function)
    
    combined_expression = random.choice(functions)
    for _ in range(num_functions - 1):
        operation = random.choices(['+', '-', '*', '/'], weights=[0.4, 0.3, 0.2, 0.1])[0]
        function = random.choice(functions)
        combined_expression = str(combined_expression) + operation + str(function)
    
    combined_expression = sympify(combined_expression)
    
    question = f"Calculez l'expression suivante : {combined_expression} = 0"

    # Ajout du solveur pour les quations polynomiales
    if combined_expression.is_polynomial(x):
        solutions = solve(Eq(combined_expression, 0), x, domain=S.Complexes)
    else :
        func_np = lambdify(x, combined_expression, modules=['numpy'])
        solutions = fsolve(func_np, 0)
        
    if len(solutions) > 0:
        question += f"\nSolutions de l'quation associe : {', '.join(str(sol) for sol in solutions)}."
    else:
        question += "\nL'quation n'a pas de solution relle."
    
    return question

# Exemple d'utilisation
print(generate_combined_expression_question(complexity=0.8, variability=0.7))
```

```python
import random
from sympy import *

def generate_sequence_question():
    x = symbols('x')
    sequence_types = ['arithmetic', 'geometric', 'quadratic']

    sequence_type = random.choices(sequence_types, weights=[0.4, 0.4, 0.2])[0]
    start = random.randint(-5, 5)
    n = random.randint(5, 10)

    if sequence_type == 'arithmetic':
        step = random.randint(1, 5)
        sequence = [start + i * step for i in range(n)]
        question = f"Donnez les {n} premiers termes de la squence arithmtique suivante : {sequence[0]}, {sequence[1]}, {sequence[2]}, ..."
        question += f"\nTermes : {', '.join(str(term) for term in sequence)}."

    elif sequence_type == 'geometric':
        ratio = random.randint(2, 5)
        sequence = [start * ratio**i for i in range(n)]
        question = f"Donnez les {n} premiers termes de la squence gomtrique suivante : {sequence[0]}, {sequence[1]}, {sequence[2]}, ..."
        question += f"\nTermes : {', '.join(str(term) for term in sequence)}."

    elif sequence_type == 'quadratic':
        a = random.randint(-5, 5)
        b = random.randint(-5, 5)
        c = random.randint(-5, 5)
        sequence = [a*x**2 + b*x + c for x in range(start, start + n)]
        question = f"Donnez les {n} premiers termes de la squence quadratique suivante : {sequence[0]}, {sequence[1]}, {sequence[2]}, ..."
        question += f"\nTermes : {', '.join(str(term) for term in sequence)}."

    return question

# Exemple d'utilisation
print(generate_sequence_question())
```

```python
import random
from sympy import *

def generate_arithmetic_expression_question():
    x, y = symbols('x y')

    # Gnration d'oprations binaires
    binary_operations = ['+', '-', '*', '/']
    num_operations = random.randint(2, 4)
    operations = random.choices(binary_operations, k=num_operations)

    # Gnration des expressions mixtes
    expressions = []
    for _ in range(num_operations + 1):
        expression_type = random.choices(['constant', 'variable', 'surds'], weights=[0.4, 0.4, 0.2])[0]

        if expression_type == 'constant':
            constant = random.randint(1, 10)
            expressions.append(constant)
        elif expression_type == 'variable':
            expressions.append(random.choice([x, y]))
        elif expression_type == 'surds':
            surd_type = random.choices(['sqrt', 'cbrt'], weights=[0.7, 0.3])[0]
            if surd_type == 'sqrt':
                expressions.append(sqrt(random.randint(1, 10)))
            elif surd_type == 'cbrt':
                expressions.append(cbrt(random.randint(1, 10)))

    # Construction de l'expression arithmtique combine
    combined_expression = expressions[0]
    question = f"Calculez l'expression suivante : {combined_expression}"

    for i in range(num_operations):
        operation = operations[i]
        expression = expressions[i + 1]
        combined_expression = str(combined_expression) + operation + str(expression)
        question += f" {operation} {latex(expression)}"

    question += " = ?"

    # Calcul de la valeur de l'expression
    try:
        value = N(simplify(combined_expression).subs({x: random.randint(1, 10), y: random.randint(1, 10)}))
        question += f"\nValeur de l'expression : {value}"
    except:
        question += "\nL'expression ne peut pas tre value."

    return question

# Exemple d'utilisation
print(generate_arithmetic_expression_question())
```

```python
import random
from sympy import *

def generate_calculus_question():
    x = symbols('x')
    
    # Gnration du type de calcul (diffrentiation, intgration ou srie)
    calculation_type = random.choices(['differentiation', 'integration', 'series'], weights=[0.4, 0.4, 0.2])[0]
    
    if calculation_type == 'differentiation':
        # Gnration d'une fonction polynomiale ou trigonomtrique
        function_type = random.choices(['polynomial', 'trigonometric'], weights=[0.7, 0.3])[0]
        
        if function_type == 'polynomial':
            degree = random.randint(1, 3)
            coeffs = [random.randint(1, 5) for _ in range(degree)]
            function = sum(coeff * x**i for i, coeff in enumerate(coeffs))
        elif function_type == 'trigonometric':
            function = random.choice([sin(x), cos(x), tan(x)])
        
        question = f"Calculez la drive de la fonction suivante : {latex(function)}."
        derivative = diff(function, x)
        answer = latex(derivative)

    elif calculation_type == 'integration':
        # Gnration d'une fonction polynomiale ou trigonomtrique
        function_type = random.choices(['polynomial', 'trigonometric'], weights=[0.7, 0.3])[0]
        
        if function_type == 'polynomial':
            degree = random.randint(1, 3)
            coeffs = [random.randint(1, 5) for _ in range(degree)]
            function = sum(coeff * x**i for i, coeff in enumerate(coeffs))
        elif function_type == 'trigonometric':
            function = random.choice([sin(x), cos(x), tan(x)])
        
        question = f"Calculez l'intgrale de la fonction suivante : {latex(function)} dx."
        integral = integrate(function, x)
        answer = latex(integral)

    elif calculation_type == 'series':
        # Gnration d'une fonction polynomiale ou trigonomtrique
        function_type = random.choices(['polynomial', 'trigonometric'], weights=[0.7, 0.3])[0]
        
        if function_type == 'polynomial':
            degree = random.randint(1, 3)
            coeffs = [random.randint(1, 5) for _ in range(degree)]
            function = sum(coeff * x**i for i, coeff in enumerate(coeffs))
        elif function_type == 'trigonometric':
            function = random.choice([sin(x), cos(x), tan(x)])

        num_terms = random.randint(2, 4)
        serie = series(function, x, 0, num_terms)
        question = f"Calculez les {num_terms} premiers termes de la srie de la fonction suivante : {latex(function)}."
        answer = latex(serie.removeO())
    
    question += f"\nRponse : {answer}"
    return question

# Exemple d'utilisation
print(generate_calculus_question())
```

```python
import random

def generate_comparison_question():
    question_type = random.choices(['closest_numbers', 'pairwise_comparisons', 'sorting'], weights=[0.4, 0.4, 0.2])[0]

    if question_type == 'closest_numbers':
        num_numbers = random.randint(3, 5)
        numbers = [random.randint(1, 100) for _ in range(num_numbers)]
        target_number = random.choice(numbers)
        closest_numbers = sorted(numbers, key=lambda x: abs(x - target_number))
        question = f"Parmi les nombres suivants, quel est le plus proche de {target_number} ?\n{', '.join(str(num) for num in numbers)}"
        answer = closest_numbers[0]

    elif question_type == 'pairwise_comparisons':
        num_numbers = random.randint(3, 5)
        numbers = [random.randint(1, 100) for _ in range(num_numbers)]
        comparison_pairs = [(random.choice(numbers), random.choice(numbers)) for _ in range(num_numbers)]
        comparisons = [f"{pair[0]} > {pair[1]}" if pair[0] > pair[1] else f"{pair[0]} < {pair[1]}" for pair in comparison_pairs]
        question = f"Parmi les paires de nombres suivantes, indiquez les comparaisons vraies :\n{', '.join(comparisons)}"
        answer = [pair for pair in comparison_pairs if pair[0] > pair[1]]

    elif question_type == 'sorting':
        num_numbers = random.randint(3, 5)
        numbers = [random.randint(1, 100) for _ in range(num_numbers)]
        sorted_numbers = sorted(numbers)
        question = f"Classez les nombres suivants dans l'ordre croissant :\n{', '.join(str(num) for num in numbers)}"
        answer = sorted_numbers

    return question, answer

# Exemple d'utilisation
question, answer = generate_comparison_question()
print(question)
print("Rponse :", answer)
```

```python
import random

def generate_measurement_question():
    question_type = random.choices(['conversion', 'working_with_time'], weights=[0.7, 0.3])[0]

    if question_type == 'conversion':
        units = ['mm', 'cm', 'm', 'km']
        from_unit = random.choice(units)
        to_unit = random.choice(units)
        value = random.randint(1, 100)
        question = f"Convertir {value} {from_unit} en {to_unit}."
        answer = convert_measurement(value, from_unit, to_unit)

    elif question_type == 'working_with_time':
        hours = random.randint(1, 12)
        minutes = random.randint(0, 59)
        question = f"Calculer le nombre de minutes coules depuis {hours}:{minutes:02d} AM jusqu' midi."
        answer = calculate_minutes(hours, minutes)

    return question, answer

def convert_measurement(value, from_unit, to_unit):
    conversion_table = {
        ('mm', 'cm'): lambda x: x / 10,
        ('mm', 'm'): lambda x: x / 1000,
        ('mm', 'km'): lambda x: x / 1000000,
        ('cm', 'mm'): lambda x: x * 10,
        ('cm', 'm'): lambda x: x / 100,
        ('cm', 'km'): lambda x: x / 100000,
        ('m', 'mm'): lambda x: x * 1000,
        ('m', 'cm'): lambda x: x * 100,
        ('m', 'km'): lambda x: x / 1000,
        ('km', 'mm'): lambda x: x * 1000000,
        ('km', 'cm'): lambda x: x * 100000,
        ('km', 'm'): lambda x: x * 1000,
    }

    conversion_func = conversion_table.get((from_unit, to_unit))
    if conversion_func:
        return conversion_func(value)
    else:
        return None

def calculate_minutes(hours, minutes):
    minutes_since_midnight = (hours * 60) + minutes
    minutes_until_noon = (12 * 60) - minutes_since_midnight
    return minutes_until_noon

# Exemple d'utilisation
question, answer = generate_measurement_question()
print(question)
print("Rponse :", answer)
```

```python
import random

def generate_number_question():
    question_type = random.choices(['base_conversion', 'remainders', 'common_divisors', 'primality', 'place_value', 'rounding_numbers'], weights=[0.3, 0.2, 0.2, 0.15, 0.1, 0.05])[0]

    if question_type == 'base_conversion':
        decimal_number = random.randint(1, 100)
        from_base = random.choice([2, 8, 10, 16])
        to_base = random.choice([2, 8, 10, 16])
        question = f"Convertir le nombre dcimal {decimal_number} en base {from_base} en base {to_base}."
        answer = convert_base(decimal_number, from_base, to_base)

    elif question_type == 'remainders':
        divisor = random.randint(2, 10)
        dividend = random.randint(11, 100)
        question = f"Calculer le reste de la division de {dividend} par {divisor}."
        answer = dividend % divisor

    elif question_type == 'common_divisors':
        num1 = random.randint(10, 50)
        num2 = random.randint(10, 50)
        divisors = [i for i in range(1, min(num1, num2) + 1) if num1 % i == 0 and num2 % i == 0]
        if len(divisors) < 2:
            return generate_number_question()  # Retry generating the question
        common_divisors = random.sample(divisors, k=random.randint(2, min(5, len(divisors))))
        question = f"Quels sont les diviseurs communs de {num1} et {num2} ?"
        answer = common_divisors

    elif question_type == 'primality':
        number = random.randint(10, 100)
        question = f"Vrifier si le nombre {number} est premier."
        answer = is_prime(number)

    elif question_type == 'place_value':
        number = random.randint(100, 999)
        digit = random.choice([1, 10, 100])
        question = f"Quelle est la valeur de position du chiffre {digit} dans le nombre {number} ?"
        answer = place_value(number, digit)

    elif question_type == 'rounding_numbers':
        decimal_number = random.uniform(1, 100)
        question = f"Arrondir le nombre dcimal {decimal_number}  l'entier le plus proche."
        answer = round(decimal_number)

    return question, answer


    return question, answer

def convert_base(decimal_number, from_base, to_base):
    if from_base == 10:
        return int(str(decimal_number), to_base)
    elif to_base == 10:
        return int(str(decimal_number), from_base)
    else:
        intermediate_base = int(str(decimal_number), from_base)
        intermediate_string = ''
        while intermediate_base > 0:
            remainder = intermediate_base % to_base
            if remainder < 10:
                intermediate_string = str(remainder) + intermediate_string
            else:
                intermediate_string = chr(ord('A') + remainder - 10) + intermediate_string
            intermediate_base //= to_base
        return int(intermediate_string, to_base)



def is_prime(number):
    if number < 2:
        return False
    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            return False
    return True

def place_value(number, digit):
    digit_str = str(digit)
    number_str = str(number)
    if digit_str in number_str:
        return number_str.index(digit_str) + 1
    else:
        return None
    
# Exemple d'utilisation
question, answer = generate_number_question()
print(question)
print("Rponse :", answer)
```

```python
import random
from sympy import symbols, Poly, expand, simplify, compose

def generate_polynomial_question():
    x = symbols('x')
    
    # Gnrer deux polynmes alatoires
    degree1 = random.randint(1, 4)
    coefficients1 = [random.randint(-5, 5) for _ in range(degree1 + 1)]
    polynomial1 = Poly(sum(coefficients1[i] * x**i for i in range(degree1)), x)
    
    degree2 = random.randint(1, 4)
    coefficients2 = [random.randint(-5, 5) for _ in range(degree2 + 1)]
    polynomial2 = Poly(sum(coefficients2[i] * x**i for i in range(degree2)), x)
    
    # Gnrer une opration alatoire
    operation = random.choice(['addition', 'simplification', 'composition', 'valuation', 'expansion'])
    
    if operation == 'addition':
        question = f"Additionnez les polynmes suivants : {polynomial1} + {polynomial2}."
        answer = simplify(polynomial1 + polynomial2, rational=True)
    elif operation == 'simplification':
        question = f"Simplifiez le polynme suivant : {polynomial1 * polynomial2}."
        answer = simplify(polynomial1 * polynomial2, rational=True)
    elif operation == 'composition':
        question = f"Composez les polynmes suivants : f(g(x)) avec f(x) = {polynomial1} et g(x) = {polynomial2}."
        answer = simplify(compose(polynomial1, polynomial2), rational=True)
    elif operation == 'valuation':
        value = random.randint(-5, 5)
        question = f"valuez le polynme suivant pour x = {value} : {polynomial1}."
        answer = polynomial1.subs(x, value)
    elif operation == 'expansion':
        question = f"Dveloppez le polynme suivant : ({polynomial1} + {polynomial2})^2."
        answer = expand((polynomial1 + polynomial2)**2)
    
    return question, answer

# Exemple d'utilisation
question, answer = generate_polynomial_question()
print(question)
print("Rponse :", answer)
```

```python
import random
from scipy.special import comb

def generate_probability_question():
    num_items = random.randint(5, 10)  # Nombre total d'lments
    num_samples = random.randint(2, 4)  # Nombre d'chantillons  prlever

    items = list(range(1, num_items + 1))

    # Gnrer l'chantillon sans remise
    sample = random.sample(items, num_samples)

    question = f"Dans un ensemble de {num_items} lments, on prleve {num_samples} lments sans remise. Quelle est la probabilit d'obtenir l'chantillon {sample} ?"

    # Calculer la probabilit
    total_possible_outcomes = comb(num_items, num_samples, exact=True)  # Nombre total de rsultats possibles
    favorable_outcomes = 1  # Il y a une seule faon d'obtenir l'chantillon prlev
    probability = favorable_outcomes / total_possible_outcomes

    answer = f"{probability:.4f}"  # Formater la rponse avec 4 dcimales

    return question, answer

# Exemple d'utilisation
question, answer = generate_probability_question()
print(question)
print("Rponse :", answer)
```

```python
import random
from sympy import symbols, residue

def generate_residue_question():
    x = symbols('x')
    functions = [1/(x**2 - 1), 1/(x*(x - 1)*(x + 1)), 1/(x**2 + 4*x + 4)]
    function = random.choice(functions)

    x0 = random.randint(-5, 5)
    res = residue(function, x, x0)

    question = f"Calculez le rsidu de la fonction {function} en x = {x0}."
    answer = f"Le rsidu de la fonction est : {res}"

    return question, answer
# Exemple d'utilisation
question, answer = generate_residue_question()
print(question)
print("Rponse :", answer)
```

```python
import random
from sympy import symbols, Eq, Function, dsolve

def generate_differential_equation_question():
    x = symbols('x')
    y = Function('y')(x)

    # Exemples d'quations diffrentielles
    equations = [Eq(y.diff(x), x**2), Eq(y.diff(x) + 2*y, x**3 - x)]

    equation = random.choice(equations)

    sol = dsolve(equation)

    question = f"Rsolvez l'quation diffrentielle suivante : {equation}"
    answer = f"La solution de l'quation est : {sol}"

    return question, answer

# Exemple d'utilisation
question, answer = generate_differential_equation_question()
print(question)
print("Rponse :", answer)
```
